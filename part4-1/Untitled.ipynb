```python
import aiohttp
import json

async with aiohttp.ClientSession(connector=aiohttp.TCPConnector(ssl=False)) as session:
    async with session.get("https://dummyjson.com/products") as resp:
        j = await resp.json()
        p = j["products"]
        print(p[2]["price"])
```
